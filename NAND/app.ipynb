```python
import tensorflow as tf;
from keras import models, layers;
import numpy as np;
import matplotlib.pyplot as plt;
```

```python
trainin_data = np.array([[0,0],[1,0],[0,1],[1,1]], "float32");

target_data = np.array([[0],[0],[0],[1]], "float32");
```

```python
model = models.Sequential([
    layers.Dense(4, input_shape=(2,)),
    layers.Dense(16, activation='relu'),
    layers.Dense(1, activation='sigmoid')
]);

model.compile(loss='mean_squared_error', optimizer='adam', metrics=['binary_accuracy']);
```

```python
history = model.fit(trainin_data, target_data, epochs=1000);

plt.plot(history.history['loss']);

plt.title('Model loss');

plt.ylabel('Loss');

plt.xlabel('Epoch');

plt.show();
```

```python
json_file_m = model.to_json();

with open("AND_model.json", "w") as json_file:
    json_file.write(json_file_m);

model.save_weights("AND_model.weights.h5");
```
