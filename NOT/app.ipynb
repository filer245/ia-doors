```python
import tensorflow as tf;
from keras import models, layers;
import numpy as np;
import matplotlib.pyplot as plt;
```

```python
training_data = np.array([[0],[1]], "float32");

target_data = np.array([[1],[0]], "float32");
```

```python
model = models.Sequential([
    layers.Dense(2, input_dim=1),
    layers.Dense(16, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])

model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
history = model.fit(training_data, target_data, epochs=800);

plt.plot(history.history['loss'])
plt.xlabel('Epocas')
plt.ylabel('Perdida')
plt.show()
```

```python
print(model.predict(np.array([1], "float32"))[0][0].round());
```

```python
model_json = model.to_json();

with open('NOT_model.json', 'w') as json_file:
    json_file.write(model_json);

model.save_weights('NOT_model.weights.h5');
```
