```python
import tensorflow as tf;
from keras import layers,models;
import numpy as np;
import matplotlib.pyplot as plt;
```

```python
training_data = np.array([[0,0], [1,0], [0,1], [1,1]], "float32");
targert_data = np.array([[0], [1], [1], [1]],"float32");
```

```python
model = models.Sequential([
    layers.Dense(1, input_shape=(2,)),
    layers.Dense(16, activation='relu'),
    layers.Dense(32, activation='relu'),
    layers.Dense(1, activation='sigmoid')
]);


model.compile(loss='binary_crossentropy', optimizer='adam');
```

```python
history = model.fit(training_data, targert_data, epochs=1000);

plt.plot(history.history['loss']);

plt.title('Model loss');   

plt.ylabel('Loss');

plt.xlabel('Epoch');

plt.show();
```

```python
print(model.predict(np.array([[1,1]], "float32")).round()[0][0]);
```

```python
json_model = model.to_json();

with open('OR_model.json', 'w') as json_file:
    json_file.write(json_model);

model.save_weights('OR_model.weights.h5');
```
