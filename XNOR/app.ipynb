```python
import tensorflow as tf;
from keras import models, layers;
import numpy as np;
import matplotlib.pyplot as plt;
```

```python
training_data = np.array([[0,0], [1,0], [0,1], [1,1]], "float32");
target_data = np.array([[1], [0], [0], [1]], "float32");
```

```python
model = models.Sequential([
    layers.Dense(4, activation='relu', input_shape=(2,)),
    layers.Dense(16, activation='relu'),
    layers.Dense(16, activation='relu'),
    layers.Dense(1, activation='sigmoid')
]);

model.compile(loss='binary_crossentropy', optimizer='adam');
```

```python
history = model.fit(training_data, target_data, epochs=1000);

plt.title("Lossed");
plt.xlabel("Epoch");
plt.ylabel("Loss");
plt.plot(history.history['loss']);

plt.show();
```

```python
print(model.predict(np.array([[1,0]], "float32")).round()[0][0]);
```

```python
json_model = model.to_json();

with open('XNOR_model.json', 'w') as json_file:
    json_file.write(json_model);

model.save_weights('XNOR_model.weights.h5');
```
