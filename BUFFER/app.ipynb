```python
import numpy as np;
import tensorflow as tf;
from keras import models, layers;
import matplotlib.pyplot as plt;
```

```python
training_data = np.array([[0],[1]], "float32");

target_data = np.array([[0],[1]], "float32");
```

```python
model = models.Sequential([
    layers.Dense(1, input_dim=1),
    layers.Dense(16, activation='relu'),
    layers.Dense(1, activation='sigmoid')
]);

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['binary_accuracy']);
```

```python
history = model.fit(training_data, target_data, epochs=500);

plt.plot(history.history['loss']);
plt.xlabel("#Epoca");
plt.ylabel("Pérdida");
plt.show();
```

```python
model_json = model.to_json();

with open("BUFFER_model.json", "w") as json_file:
    json_file.write(model_json);

model.save_weights("BUFFER_model.weights.h5");
```
